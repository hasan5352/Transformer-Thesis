```python
def calculate_mCE(baseline_errors, model_errors):
    """
    Calculate mean Corruption Error (mCE).

    Args:
        baseline_errors (list or array): baseline error rates per corruption type
        model_errors (list or array): model error rates per corruption type

    Returns:
        float: mCE value
    """
    ratios = [m / b for m, b in zip(model_errors, baseline_errors)]
    mCE = sum(ratios) / len(ratios)
    return mCE
```
